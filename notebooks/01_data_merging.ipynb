## Merge Tourists

```
import io

inbound_annual = pd.read_csv('Number-of-Inbound-Tourists-by-Main-Purpose-of-Trips.csv')
inbound_monthly = pd.read_csv('Number-of-Inbound-Tourists-by-Main-Purpose-of-Trips.minth.csv')
domestic_annual = pd.read_csv('Number-of-Domestic-Tourists-by-Main-Purpose-of-Trips.csv')
domestic_monthly = pd.read_csv('Number-of-Domestic-Tourists-by-Main-Purpose-of-Trips.month.csv')

for df in [inbound_annual, domestic_annual]:
    df.rename(columns={df.columns[0]: 'Period'}, inplace=True)
    df['Frequency'] = 'Annual'
for df in [inbound_monthly, domestic_monthly]:
    df.rename(columns={df.columns[0]: 'Period'}, inplace=True)
    df['Frequency'] = 'Monthly'

tourists_combined = pd.concat([inbound_annual, inbound_monthly,
                                domestic_annual, domestic_monthly],
                               ignore_index=True)
print(f'Tourists Combined: {tourists_combined.shape}')
tourists_combined.head()
```

## Merge Overnight Stays

```
overnight_annual = pd.read_csv('Annual-Evolution-of-the-Number-of-Overnight-Stays-by-Type-of-Tourist.csv')
overnight_monthly = pd.read_csv('Monthly-Evolution-of-the-Number-of-Overnight-Stays-by-Type-of-Tourist.csv')

overnight_annual.rename(columns={overnight_annual.columns[0]: 'Period'}, inplace=True)
overnight_monthly.rename(columns={overnight_monthly.columns[0]: 'Period'}, inplace=True)
overnight_annual['Frequency'] = 'Annual'
overnight_monthly['Frequency'] = 'Monthly'

overnight_combined = pd.concat([overnight_annual, overnight_monthly], ignore_index=True)
print(f'Overnight Combined: {overnight_combined.shape}')
```

## Merge KPI Indicators

```
length = pd.read_csv('Monthly-Evolution-of-the-Average-Length-of-Stay-acordding-to-Type-of-Tourist.csv')
spending_trip = pd.read_csv('Monthly-Evolution-of-the-Average-Spending-per-Trip-according-to-Type-of-Tourist.csv')
spending_night = pd.read_csv('Monthly-Evolution-of-the-Average-Spending-per-Night-according-to-Type-of-Tourist.csv')

kpi_combined = length.merge(spending_trip, on=['Month','Type of tourist'])\
                      .merge(spending_night, on=['Month','Type of tourist'])
print(f'KPI Combined: {kpi_combined.shape}')
```

## Merge Expenditure

```
exp_inbound = pd.read_csv('Annual-Expenditure-on-Inbound-Tourist-Trips-by-Purpose-of-Visit.csv')
exp_domestic = pd.read_csv('Annual-Expenditure-on-Domestic-Tourist-Trips-by-Purpose-of-Visit.csv')

exp_inbound['Type of tourist'] = 'Inbound'
exp_domestic['Type of tourist'] = 'Domestic'

expenditure_combined = pd.concat([exp_inbound, exp_domestic], ignore_index=True)
print(f'Expenditure Combined: {expenditure_combined.shape}')
```

## Save to Drive

```
save_path = '/content/drive/MyDrive/Saudi_Tourism_Project/data/clean'
os.makedirs(save_path, exist_ok=True)

tourists_combined.to_csv(f'{save_path}/01_Tourists_Combined.csv', index=False)
overnight_combined.to_csv(f'{save_path}/02_Overnight_Combined.csv', index=False)
kpi_combined.to_csv(f'{save_path}/03_KPI_Combined.csv', index=False)
expenditure_combined.to_csv(f'{save_path}/04_Expenditure_Combined.csv', index=False)

print('All files saved to Drive ✅')
```
