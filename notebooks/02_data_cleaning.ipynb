## Setup

```
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import os

base = '/content/drive/MyDrive/Saudi_Tourism_Project/data/clean'
print('Setup done ✅')
```

## Load & Inspect

```
files = {
    '01_Tourists': f'{base}/01_Tourists_Combined.csv',
    '02_Overnight': f'{base}/02_Overnight_Combined.csv',
    '03_KPI': f'{base}/03_KPI_Combined.csv',
    '04_Expenditure': f'{base}/04_Expenditure_Combined.csv'
}

for name, path in files.items():
    df = pd.read_csv(path)
    print(f'\n{name}: {df.shape}')
    print(f'Missing: {df.isnull().sum().sum()}')
    print(f'Duplicates: {df.duplicated().sum()}')
```

## Clean Data

```
# Tourists
tourists = pd.read_csv(f'{base}/01_Tourists_Combined.csv')
tourists_clean = tourists.drop_duplicates()

# KPI - Fix Date Format
kpi = pd.read_csv(f'{base}/03_KPI_Combined.csv')
kpi['Month'] = kpi['Month'].str.replace('--', '-')

# Overnight
overnight = pd.read_csv(f'{base}/02_Overnight_Combined.csv')

# Expenditure - Convert to Billions
exp = pd.read_csv(f'{base}/04_Expenditure_Combined.csv')
exp['Expenditure_Billions_SAR'] = (exp['Expenditure'] / 1e9).round(2)
exp = exp.drop(columns=['Expenditure'])

print(f'Tourists: {len(tourists)} → {len(tourists_clean)} rows ✅')
print(f'KPI Date fixed ✅')
print(f'Expenditure converted to Billions ✅')
```

## Save Clean Files

```
tourists_clean.to_csv(f'{base}/01_Tourists_CLEAN.csv', index=False)
overnight.to_csv(f'{base}/02_Overnight_CLEAN.csv', index=False)
kpi.to_csv(f'{base}/03_KPI_CLEAN.csv', index=False)
exp.to_csv(f'{base}/04_Expenditure_CLEAN.csv', index=False)

print('All clean files saved ✅')
```
